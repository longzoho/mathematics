```python
# Khai báo biến
x, y = symbols('x y')

# Định nghĩa hàm số
y1 = x ** 2
y2 = 2 - x

# Giải hệ phương trình để tìm điểm giao nhau
intersection_points = solve([y1 - y2, y - y1], (x, y))

# Vẽ đồ thị
p1 = plot(y1, y2, (x, -5, 5), show=False, legend=False)

# Vẽ điểm giao nhau canh với trục tung
for point in intersection_points:
    p1.append(plot(point, point.subs(x, point), marker='.', show=False,
                   label=f'({point.evalf()}, {y1.subs(x, point.evalf()).evalf()})')[1])

p1.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Khai báo biến ký hiệu
x = sp.Symbol('x')

# Tìm điểm giao nhau bằng cách giải phương trình
intersection_points = sp.solve(x ** 2 - (2 - x), x)

# Tách ra các giá trị x và y tương ứng
x_values = [point.evalf() for point in intersection_points]
y_values = [(2 - point).evalf() for point in intersection_points]

# Vẽ đồ thị
x_range = np.linspace(-3, 3, 400)
y1_values = x_range ** 2
y2_values = 2 - x_range

plt.plot(x_range, y1_values, label='$y = x^2$')
plt.plot(x_range, y2_values, label='$y = 2 - x$')
plt.scatter(x_values, y_values, color='red', label='Các điểm giao nhau')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Đồ thị của hai hàm số và các điểm giao nhau')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sympy import symbols, plot, Eq, solve
from sympy.plotting import plot

x = symbols('x')

# Định nghĩa các hàm số
f = x ** 2
g = x + 1

# Tìm tọa độ giao điểm
solution = solve(Eq(f, g), x)
x_intersect = solution[0]
y_intersect = f.subs(x, x_intersect)

# Vẽ đồ thị f và g
p1 = plot(f, (x, -5, 5), show=False, line_color='blue', legend=True, label='f(x)')
p2 = plot(g, (x, -5, 5), show=False, line_color='red', legend=True, label='g(x)')

# Vẽ điểm giao điểm
p3 = plot(y_intersect, (x, x_intersect, x_intersect), show=False, marker='o', markersize=8, color='black',
          label='Giao điểm')

# Thêm đồ thị g và điểm giao điểm vào đồ thị f
p1.append(p2[0])
p1.append(p3[0])

# Đặt tiêu đề và nhãn cho hệ trục
p1.title = 'Đồ thị hai hàm số và giao điểm'
p1.xlabel = 'x'
p1.ylabel = 'y'

# Hiển thị đồ thị
p1.show()
```

```python
from sympy import symbols, plot, Eq, solve
from sympy.plotting import plot
import matplotlib.pyplot as plt

x = symbols('x')

# Định nghĩa các hàm số
f = x ** 2
g = 2 - x

# Tìm tọa độ giao điểm
solution = solve(Eq(f, g), x)
x_intersect = solution[0]
y_intersect = f.subs(x, x_intersect)

# Vẽ đồ thị f và g
p1 = plot(f, (x, -5, 5), show=False, line_color='blue', label='f(x)')
p2 = plot(g, (x, -5, 5), show=False, line_color='red', label='g(x)')

# Thêm đồ thị g và điểm giao điểm vào đồ thị f
p1.append(p2[0])

# Đặt tiêu đề và nhãn cho hệ trục
p1.title = 'Đồ thị hai hàm số và giao điểm'
p1.xlabel = 'x'
p1.ylabel = 'y'

# Hiển thị tọa độ giao điểm bằng matplotlib.pyplot.text
text = plt.text(x_intersect, y_intersect, f"({x_intersect:.2f}, {y_intersect:.2f})", fontsize=12, ha='right',
                va='bottom')
p1.append(text)
# Hiển thị lại biểu đồ
plt.show()
```
